```python
# 8.
a = np.arange(0, 12)
A = np.reshape(a, (4, 3))
At = A.T
B = np.dot(A, At)
np.linalg.det(B)
# Обратную матрицу вычислить нельзя, т.к. определитель равен нулю
```

```python
# 9.
np.random.seed(42)
c = np.arange(0, 16)
C = c.reshape(4, 4)
D = B + (np.dot(C, 10))
D_det = np.linalg.det(D)
D_rank = np.linalg.matrix_rank(D)
D_inv = np.linalg.inv(D)
print(D_det, D_rank, D_inv)
```

```python
# 10.
D_inv[D_inv < 0] = 0
D_inv[D_inv > 0] = 1

E = np.where(D_inv, B, C)
print('D_INV \n', D_inv, '\n')
print('B \n', B, '\n')
print('C \n', C, '\n')
print('E \n', E, '\n')
```
